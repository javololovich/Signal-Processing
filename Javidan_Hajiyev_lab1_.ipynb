Question 4- Edit the code inorder to represent the signal w, within one second, after shift in time by π/3, and a multiplication by 2. The frequency of the new signal has to be set to 5. Choose the adaquate time step in order to have a smooth curve.

```python
import numpy as np
import matplotlib.pyplot as plot

f_new = 5
n = np.arange(0, 1, 0.003)
w_values_modified = 2 * np.cos(2 * np.pi * f_new * (n - np.pi/3))
plot.figure(figsize=(10, 5))
plot.plot(n, w_values_modified, label="Modified cos")
plot.xlabel('$n$')
plot.ylabel('$w(t)$')
plot.title('Modified Signal $w(t) = 2\cos(2\pi f_{\mathrm{new}} (t - \pi/3))$')
plot.legend()
plot.grid(True)
plot.show()
```

Question 5- represent the damped signal Capture d’écran 2023-10-03 à 17.13.30.png

A=2, a=0.5, f=2, step time=0,05.


```python
import numpy as np
import matplotlib.pyplot as plt
A = 2
a = 0.5
f = 2
time_step = 0.05
t = np.arange(0, 10, time_step)
y_values = A * np.exp(-a * t) * np.cos(2 * np.pi * f * t)
plt.figure(figsize=(10, 5))
plt.plot(t, y_values, label="Damped Signal")
plt.xlabel('$t$')
plt.ylabel('$y(t)$')
plt.title(' $y(t) = A \cdot e^{-at} \cdot \cos(2\pi ft)$')
plt.legend()
plt.grid(True)
plt.show()
```

Question 7- Define, within -3 and 3,  the discrete function z[n] = A rect(n) with time step egal to 0.1.
(You can use the function 'where' of the numpy library)

```python
import numpy as np
import matplotlib.pyplot as plt
n = np.arange(-3, 3.0, 0.1)
A = 2
z_values = A * np.where(np.logical_and(n >= -1, n <= 1), 1, 0)
plt.plot(n, z_values)
plt.xlabel('$n$')
plt.ylabel('$z[n]$')
plt.title('Discrete Function $z[n] = A \cdot \mathrm{rect}(n)$')
plt.legend()
plt.grid(True)
plt.show()


```

Question 9- Compute the convolution product of a(t) by z(t) in the intervall [-3,3] with time step egal to 0.1.

---


(You can use the function convolve of numpy library with mode="same")

```python
import numpy as np
import matplotlib.pyplot as plt
t = np.arange(-3, 3, 0.1)
a_values = np.where(np.logical_and(t >= -0.5, t <= 0.5), 0.5 - t, 0)
z_values = np.where(np.logical_and(t >= -0.5, t <= 0.5), 1, 0)
output = np.convolve(a_values, z_values, mode="same")
plt.plot(t, out)
plt.xlabel('$t$')
plt.ylabel('Convolution Product')
plt.title('Convolution of a(t) and z(t)')
plt.legend()
plt.grid(True)
plt.show()
```
