Let x the analog signal defined as: x(t) = [sinc((t-t0)/T)]^2

Question 1- Plot the signal x for t0 = 3, T=2 in the interval [-2,8] with range 0.1


```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(-2, 8, 0.1)
t0 = 3
T = 2
x = (np.sinc((t-t0)/T))**2

plt.plot(t, x)
plt.title('Signal Plot')
plt.xlabel('Time (t)')
plt.ylabel('x(t)')
plt.grid(True)
plt.show()
```

Question 2- Calculate the Fourier Transform of x using fft function then plot the module of the result

```python


X = np.fft.fft(x)
X = np.fft.fftshift(X)

N = len(t)
dt = 0.1
frequencies = np.fft.fftfreq(N, dt)
frequencies = np.fft.fftshift(frequencies)


plt.stem(frequencies, np.abs(X))
plt.title('Magnitude of the Fourier Transform of x(t)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('|X(f)|')
plt.grid(True)
plt.show()
```

Question 4- Calculate the energy of the signal by implementing the definitions

```python

x = np.sinc((t - t0) / T) ** 2
energy = np.trapz(np.abs(x)**2, t)

print("Energy of the signal: ", energy)
```
