Question 2- Calculate the fourier transform of x(t) using fft function then compare the result to theorical one (calculated anatically in the previous question)

```python
import numpy as np
import matplotlib.pyplot as plt

f0 = 2
A = 3
duration = 1

t = np.linspace(0, duration, num=1000, endpoint=False)

x = A * np.cos(2 * np.pi * f0 * t)

X = np.fft.fft(x)
frequencies = np.fft.fftfreq(len(t), d=t[1] - t[0])

plt.subplot(2, 1, 1)
plt.plot(t, x)
plt.title('Original Signal x(t) = cos(2πf0t)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

plt.subplot(2, 1, 2)
plt.plot(frequencies, np.abs(X))
plt.title('Fourier Transform of x(t)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')

plt.tight_layout()
plt.show()
```

Question 6- Calculate the fourier transform of z(t) using fft function then compare the result to theorical one (calculated anatically in the previous question)

```python
import numpy as np
import matplotlib.pyplot as plt

f0 = 2
A = 3
duration = 1

t = np.linspace(0, duration, num=1000, endpoint=False)

z = A * np.exp(1j * 2 * np.pi * f0 * t)

Z = np.fft.fft(z)
frequencies = np.fft.fftfreq(len(t), d=t[1] - t[0])

plt.subplot(2, 1, 1)
plt.plot(t, np.real(z))
plt.title('Real part of z(t) = exp(j2πf0t)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

plt.subplot(2, 1, 2)
plt.plot(frequencies, np.abs(Z))
plt.title('Fourier Transform of z(t)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')

plt.tight_layout()
plt.show()
```
