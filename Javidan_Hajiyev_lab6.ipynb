In this lab, we propose to calculate the Fourier transform of s(t)= (sin(2πt))^2 = 1/2 * ( 1 - cos(4πt))


Question 1- Plot in the same figure, the following functions: sin(2πt), cos(2πt), -1/2 * cos(4πt) and (sin(2πt))^2 for t in [0,1]

```python
import numpy as np
import matplotlib.pyplot as plt
t = np.linspace(0, 1, 200)
f1 = np.sin(2*np.pi*t)
f2 = np.cos(2*np.pi*t)
f3 = -0.5 * np.cos(4*np.pi*t)
f4 = (np.sin(2*np.pi*t))**2
plt.figure(figsize=(10, 6))
plt.plot(t, f1, label='sin(2πt)')
plt.plot(t, f2, label='cos(2πt)')
plt.plot(t, f3, label='-1/2 cos(4πt)')
plt.plot(t, f4, label='(sin(2πt))^2')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.legend()
plt.show()

```

Question 4-Calculate the Fourier transform Sf using fft, then plot Hf = j * Sf

```python
import numpy as np
import matplotlib.pyplot as plt
t = np.linspace(0, 1, 50, endpoint=False)
s_t = np.sin(2 * np.pi * t)
S_f = np.fft.fft(s_t)
freq = np.fft.fftfreq(len(t), d=t[1] - t[0])
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.stem(freq, np.abs(S_f))
plt.xlabel('Frequency (Hz)')
plt.ylabel('|S(f)|')
plt.title('Magnitude of Fourier Transform S(f)')
```

Question 6- Calculate then plot the Fourier transform of s(t)

```python
t = np.linspace(0, 1, 50)
s_t = 0.5 * (1 - np.cos(4*np.pi*t))
s_f = np.fft.fft(s_t)
w = np.fft.fftfreq(len(t))
plt.figure(figsize=(12, 6))
plt.subplot(122)
plt.stem(w, np.abs(s_f))
plt.title('Fourier Transform')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')
plt.tight_layout()
plt.show()
```
